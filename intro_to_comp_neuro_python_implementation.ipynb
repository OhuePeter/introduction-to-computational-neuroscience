### Euler's Method

```python
# Parameters
tau = 10  # ms
R = 1     # MΩ
V_rest = -65  # mV
V_reset = -70 # mV
V_thresh = -50 # mV
I = 1.5  # nA constant input
dt = 0.1
time = np.arange(0, 100, dt)
```

```python
# Initialization
V = np.zeros_like(time)
V[0] = V_rest
```

```python
# Simulation
for i in range(1, len(time)):
    dV = (-(V[i-1] - V_rest) + R * I) * (dt / tau)
    V[i] = V[i-1] + dV
    if V[i] >= V_thresh:
        V[i] = V_reset  # spike and reset
```

```python
# Plot
plt.figure(figsize=(10, 4))
plt.plot(time, V)
plt.title('Leaky Integrate-and-Fire Neuron')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.grid(True)
plt.show()
```

### Simulation of a simplified neuron

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
V_rest = -65       # Resting membrane potential (mV)
V_th = -50         # Threshold potential (mV)
V_reset = -65      # Reset potential (mV)
tau = 10           # Membrane time constant (ms)
R = 1              # Membrane resistance (MΩ)
dt = 0.1           # Time step (ms)
time = np.arange(0, 100, dt)  # Simulation time
```

```python
# External input current
I = 1.5 * np.ones_like(time)
```

```python
# Initialize membrane potential
V = np.zeros_like(time)
V[0] = V_rest
```

```python
# Simulation
spikes = []
for t in range(1, len(time)):
    dV = (-(V[t-1] - V_rest) + R * I[t-1]) / tau
    V[t] = V[t-1] + dV * dt

    if V[t] >= V_th:
        V[t] = V_reset
        spikes.append(time[t])
```

```python
# Plotting
plt.figure(figsize=(10, 4))
plt.plot(time, V, label='Membrane Potential')
plt.scatter(spikes, [V_th]*len(spikes), color='red', label='Spikes')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Leaky Integrate-and-Fire Neuron Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

### Python Implementation of Synaptic Input

```python
import numpy as np
import matplotlib.pyplot as plt

def psp(t, t_spike, tau, weight):
    """Compute postsynaptic potential for a given spike time"""
    response = np.zeros_like(t)
    response[t >= t_spike] = weight * np.exp(-(t[t >= t_spike] - t_spike)/tau)
    return response
```

```python
# Time array
t = np.linspace(0, 100, 1000)
```

```python
# Simulate PSP for three spikes
psp_total = np.zeros_like(t)
spike_times = [20, 50, 80]
for t_spike in spike_times:
    psp_total += psp(t, t_spike, tau=10, weight=1.0)

plt.plot(t, psp_total)
plt.title('Summed Postsynaptic Potentials from Multiple Spikes')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential')
plt.grid(True)
plt.show()
```

### Entropy

### Simulating STDP in Python

- To bring STDP to life, we simulate a pair of neurons undergoing synaptic adjustment based on spike events:


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A_plus, A_minus = 0.01, 0.012
tau_plus, tau_minus = 20, 20
w = 0.5  # initial weight

pre_spikes = np.array([20, 70, 150])  # ms
post_spikes = np.array([25, 75, 145])  # ms

for t_pre in pre_spikes:
    for t_post in post_spikes:
        dt = t_post - t_pre
        if dt > 0:
            w += A_plus * np.exp(-dt / tau_plus)
        elif dt < 0:
            w -= A_minus * np.exp(dt / tau_minus)

print(f"Updated synaptic weight: {w:.4f}")
```

### Rate-Based Learning Rules

```python
eta = 0.001
x_pre = 5  # Hz
y_post = 3  # Hz
w = 0.2
w += eta * x_pre * y_post
print(f"Updated weight with Hebbian learning: {w:.4f}")
```

### Bayesian fusion of visual and auditory estimates

```python
def bayesian_integration(mu_v, sigma_v, mu_a, sigma_a):
    weight_v = 1 / sigma_v**2
    weight_a = 1 / sigma_a**2
    mu_combined = (mu_v * weight_v + mu_a * weight_a) / (weight_v + weight_a)
    sigma_combined = 1 / np.sqrt(weight_v + weight_a)
    return mu_combined, sigma_combined
```

### Writing a Leaky Integrate-and-Fire Model in Brian2

```python
from brian2 import *

start_scope()

# Define neuron equations
tau = 10*ms
eqs = '''
dv/dt = (I - v) / tau : 1
I : 1
'''

# Create neuron group
G = NeuronGroup(1, eqs, threshold='v > 1', reset='v = 0', method='exact')
G.I = 1.2

# Record voltage
M = StateMonitor(G, 'v', record=True)
sp = SpikeMonitor(G)

run(100*ms)

plot(M.t/ms, M.v[0])
xlabel('Time (ms)'); ylabel('v')
title('Brian2 Simulation: Single LIF Neuron')
show()
```

**Extending to multiple neurons and synapses is straightforward. For example, a recurrent network of 100 excitatory neurons:**

```python
start_scope()
N = 100
G = NeuronGroup(N, eqs, threshold='v > 1', reset='v = 0', method='exact')
G.I = '0.5 + 0.5*rand()'
S = Synapses(G, G, on_pre='v_post += 0.2')
S.connect(p=0.1)

M = StateMonitor(G, 'v', record=0)
sp = SpikeMonitor(G)

run(500*ms)

plot(M.t/ms, M.v[0])
xlabel('Time (ms)'); ylabel('v')
title('Spiking Activity of Neuron 0')
show()
```
